## 3) Menghitung jarak euclidean

Tahapan awal dari algoritma KNN adalah perhitungan jarak. Salah satu metode perhitungan jarak yang bisa digunakan adalah jarak Euclidean. Buatlah fungsi bernama **jarakEu** yang berfungsi menghitung jarak euclidean dari dua buah vektor (tanpa kelas)

```python
import numpy as np

def euclidianDistance(data1, data2):
  distance = np.square(data1 - data2)
  distance = np.sum(distance)
  return np.sqrt(distance)
```

Uji fungsi **jarakEu** untuk menghitung jarak antara data latih pertama dengan data uji pertama.

```python
distance = euclidianDistance(df_train.iloc[0], df_test.iloc[0])
print(distance)
```

## 4) Algoritma KNN

Implementasikan algoritma KNN dengan tahapan-tahapan berikut :
1. Hitung jarak antara data uji dengan setiap data latih.
2. Cari *k* data latih dengan jarak terkecil.
3. Tentukan kelas dari *k* data latih tersebut.
4. Kelas data uji ditentukan dari mayoritas kelas *k* data latih.

Buatlah fungsi dengan nama **knn** yang menerima input berupa *k*, sekumpulan data latih dan labelnya, serta sebuah data uji

```python
from collections import Counter

def knn(k, data_train, label_train, data_test):
  distances = np.array([euclidianDistance(data_train.iloc[x], data_test) for x in range (data_train.shape[0])])
  print(distances)
  index_k_minumum = distances.argsort()[:k]
  print(index_k_minumum)
  k_labels = label_train.iloc[index_k_minumum].to_numpy()
  print(k_labels)
  counter = Counter(k_labels)
  label_test = counter.most_common(1)[0][0]
  return label_test
```

Lakukan pengujian fungsi **knn** untuk menentukan kelas dari data uji pertama

```python
result = knn(3, df_train, label_train, df_test.iloc[0])
print(result)
```

```python
#jawaban soal 1
def jarakCity(data1, data2):
  return np.sum(np.abs(data1 - data2))
```

```python
#jawaban soal 2
def knn_multi(k, data_train, label_train, data_test):
  result = []
  for i in range (data_test.shape[0]):    
    distances = np.array([jarakCity(data_train.iloc[x], data_test.iloc[i]) for x in range(data_train.shape[0])])
    index_k_minumum = distances.argsort()[:k]
    k_labels = label_train.iloc[index_k_minumum].to_numpy()
    counter = Counter(k_labels)
    label_test = counter.most_common(1)[0][0]
    result.append(label_test)
    
  return pd.DataFrame(result)[0]
```

```python
#jawaban soal 3
def cek_hasil(label_test, label_result):
  if label_test.shape[0] != label_result.shape[0]:
    return KeyError("Ukuran label_predict tidak sama dengan ukuran label_test")
  
  counter = 0
  for i in range(label_test.shape[0]):
    if label_test.iloc[i] == label_result[i]:
      counter += 1
  
  return counter
```

```python
#jawaban soal 4
label_predict = knn_multi(3, df_train, label_train, df_test)
print(label_predict)
```

## Penjelasan

Berdasarkan hasil di atas didapatkan jumlah label  antara label hasil prediksi dan label yang sudah ada berbeda. Hal ini terjadi karena karakteristik dari algoritma KNN itu sendiri, dimana KNN menentukan label dari suatu data yang akan di prediksi berdasarkan jumlah kategori terbanyak dari data yang dekat dengan data prediksi tersebut. Jumlah data terdekat yang diambil akan sejumlah K (hyperparameter), dimana K ini merepresentasikan jumlah data terdekat yang akan diambil.

Berdasarkan karakteristik tersebut, jumlah label yang sama berbeda dimana nilai K=3 memiliki 30 label yang sama sementara K=7,K=9,K=13 memiliki 29 label yang sama. Hal ini terjadi karena jangkauan (K) dari KNN tersebut berbeda-beda menyebabkan jumlah data terdekat yang digunakan juga berbeda sehingga akan mempengaruhi penentuan atau klasifikasi dari label data prediksi.

Contoh :
K = 3
Dengan nilai K diatas terdapat 2 data dengan label A dan 1 data dengan label B sehingga data prediksi akan berlabel A

K = 13
Dengan nilai K diatas terdapat 4 data dengan label A, 4 data dengan label B, 5 data dengan label C sehingga data prediksi akan berlabel C.

Oleh karena itu diperlukan penentuan nilai K yang tepat agar menghasilkan data prediksi yang optimal. Terdapat beberapa metode seperti hyperparameter tuning, elbow method, nilai k harus ganjil, dan masih banyak lagi.
