Buatlah plot bertipe 'line' untuk masing-masing variabelnya.
- Plot line yang menunjukkan semua variabel dimana setiap variabel memiliki 1 figure

```python
var = ['RI', 'Na', 'Mg', 'Al', 'Si', 'K', 'Ca', 'Ba', 'Fe']

for v in var:
    lines = plt.plot(df.index, df[v])
    plt.setp(lines, color='b', linewidth=2.0)
    plt.legend(v)
    plt.xlabel('Index')
    plt.ylabel(v)
    plt.setp(lines, 'color', 'b', 'linewidth', 2.0)
    plt.show()
```

Buatlah plot bertipe 'line' untuk masing-masing variabelnya.
- Plot line yang menunjukkan semua variabel yang di grouping berdasarkan typenya

```python
colors = {
    "'build wind float'": 'blue',
    "'vehic wind float'": 'green',
    'tableware': 'red',
    "'build wind non-float'": 'purple',
    'headlamps': 'orange'
}


for column in df.columns[:-1]: 
    for category, color in colors.items():
        plt.plot(df[df['Type'] == category][column], label=category, color=color)
    plt.title(f'Line Plot for {column}')
    plt.xlabel('Index')
    plt.ylabel(column)
    plt.legend()
    plt.grid(True)
    plt.show()
```
